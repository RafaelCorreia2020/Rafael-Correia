**Passo a passo de solução do desafio**



```python
# lógica de programação

# Escreva sempre o passo a passo em português primeiro paa depois começar a criar o programa

# passo 0 :  entender o desafio que você quer resolver

# passo 1 : percorrer todos os arquivos da nossa base de dados (pasta vendas)

import os #bibloteca para mexer com arquivos e pastas do pc ou do drive.
import pandas as pd

lista_arquivo = os.listdir("/content/drive/MyDrive/Python/Vendas")
display(lista_arquivo)


```

```python
tabela_total = pd.DataFrame() #uma tabela vazia no python que vai preenchendo

# passo 2 : Importar as bases de dados de vendas
for arquivo in lista_arquivo:
# se tem "vendas" no nome do aquivo, então (função if)
  if "Vendas" in arquivo:
    # importar o arquivo (base de dados)
    tabela = pd.read_csv(f"/content/drive/MyDrive/Python/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela)

# passo 3 : Tratar/compilar  as bases de dados (nem sempre está pronta)

display(tabela_total)

```

```python
# passo 4 : calcular o produto mais vendido (em quantidade)
#precisamos agrupar os produtos e somar as quantidade vendida de cada um deles
tabela_produtos = tabela_total.groupby('Produto').sum()
#feito pra selecionar a coluna que queriamos e colocar em ordem decrescente (false)
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
display(tabela_produtos)
```

```python
# passo 5 : calcular o produto que mais faturou (em faturamento)
#precisamos criar uma coluna que faça a operação de multiplicação da quantidade vendida pelo preço unitário.
tabela_total ['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']

tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_faturamento)
```

```python
# passo 6 : calcular a loja/cidade que mais vendeu (em faturamento)- criar gráfico/dashboard

tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_lojas)
```

```python
#vamos criar um gráfico. 3 pacotes sao muito usados : plotly,
import plotly.express as px
#cria grafico
grafico = px.bar(tabela_lojas, x= tabela_lojas.index, y='Faturamento')
#exibe grafico
grafico.show()
```
